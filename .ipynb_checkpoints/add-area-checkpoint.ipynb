```python
#para ver areas de cobertura usar el boton cobertus de la version local del app cargar el shapefile en jupyter es imposible




import os
import glob
import geopandas as gpd
import pandas as pd

BASE_FILE = "calles_bsas_fixed.geojson"
TO_ADD_DIR = "to_add"

def load_geojson_verbose(path: str) -> gpd.GeoDataFrame:
    print(f"\n[LOAD] {path}")
    gdf = gpd.read_file(path)
    print(f"  -> rows: {len(gdf):,}")
    print(f"  -> columns: {list(gdf.columns)}")
    print(f"  -> geom types: {gdf.geometry.geom_type.value_counts(dropna=False).to_dict()}")
    print(f"  -> crs: {gdf.crs}")
    return gdf

# 1) Load base
if not os.path.exists(BASE_FILE):
    raise FileNotFoundError(f"Base file not found: {BASE_FILE}")

base = load_geojson_verbose(BASE_FILE)

# 2) Find files to add
patterns = [os.path.join(TO_ADD_DIR, "*.geojson"), os.path.join(TO_ADD_DIR, "*.json")]
add_files = []
for p in patterns:
    add_files.extend(glob.glob(p))

add_files = sorted(add_files)
print(f"\n[SCAN] Found {len(add_files)} files in ./{TO_ADD_DIR}/")
for f in add_files[:20]:
    print("  -", f)
if len(add_files) > 20:
    print(f"  ... (+{len(add_files)-20} more)")

if not add_files:
    print("\nNothing to add. Exiting without changes.")
else:
    # 3) Load to_add files (align CRS to base if needed)
    gdfs = [base]
    for f in add_files:
        g = load_geojson_verbose(f)

        # If CRS differs but both are known, reproject to base CRS
        if base.crs is not None and g.crs is not None and base.crs != g.crs:
            print(f"  [CRS] Reprojecting {f} from {g.crs} -> {base.crs}")
            g = g.to_crs(base.crs)

        # If base has CRS and added file doesn't, assume it matches (common with Overpass exports)
        if base.crs is not None and g.crs is None:
            print(f"  [CRS] {f} has no CRS. Assuming it matches base: {base.crs}")
            g = g.set_crs(base.crs, allow_override=True)

        gdfs.append(g)

    # 4) Merge (pandas concat keeps all columns; missing columns become NaN)
    print("\n[MERGE] Concatenating GeoDataFrames...")
    merged = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True), crs=base.crs)
    print(f"  -> merged rows: {len(merged):,}")

    # 5) Optional: keep only line geometries (streets should be LineString/MultiLineString)
    print("\n[FILTER] Keeping only line geometries (LineString / MultiLineString)...")
    is_line = merged.geometry.geom_type.isin(["LineString", "MultiLineString"])
    before = len(merged)
    merged = merged[is_line].copy()
    print(f"  -> kept {len(merged):,} of {before:,}")

    # 6) Optional: drop exact duplicates by geometry (can be expensive on huge files; but often helps)
    print("\n[DEDUP] Dropping exact-duplicate geometries (may take a bit)...")
    merged["_wkb"] = merged.geometry.apply(lambda g: g.wkb_hex if g is not None else None)
    before = len(merged)
    merged = merged.drop_duplicates(subset=["_wkb"]).drop(columns=["_wkb"])
    print(f"  -> removed {before - len(merged):,} duplicates, now {len(merged):,} rows")

    # 7) Write safely (temp -> replace)
    tmp_file = BASE_FILE + ".tmp.geojson"
    print(f"\n[WRITE] Writing temp: {tmp_file}")
    merged.to_file(tmp_file, driver="GeoJSON")
    print(f"[WRITE] Overwriting: {BASE_FILE}")
    os.replace(tmp_file, BASE_FILE)

    print("\n✅ Done. calles_bsas_fixed.geojson updated.")
```
